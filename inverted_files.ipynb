Implement a program for retrieval of documents using inverted files.

```python
import re
#Imports regular expression module, used to remove punctuation and extract words cleanly.

# Define the documents
document1 = "The quick brown fox jumped over the lazy dog."
document2 = "The lazy dog slept in the sun."
```

```python
# Step 1: Preprocessing - lowercase + remove punctuation + tokenize
tokens1 = re.findall(r'\w+', document1.lower())  #document1.lower() → converts text into lowercase
tokens2 = re.findall(r'\w+', document2.lower())  #re.findall(r'\w+', ...) → extracts only words, removes punctuation like .

# Combine unique terms
terms = set(tokens1 + tokens2)
# Combines both token lists and removes duplicates using set()
# Now terms contains all unique words from both documents.
```

```python
# Step 2: Build inverted index
inverted_index = {}  #Creates an empty dictionary to store the index

for term in terms:         #Loops through every unique word one by one.
    docs = []              #Creates an empty list to store document names that contain the current word.
    if term in tokens1:              #Checks if the word exists in doc1
        docs.append("Document 1")
    if term in tokens2:
        docs.append("Document 2")
    inverted_index[term] = docs        #Stores the word + list of documents in dictionary.
```

```python
# Step 3: Print inverted index
print("Inverted Index:\n")
for term, docs in inverted_index.items():
    print(term, "→", ", ".join(docs))       #Prints each term and the documents where it exists & ", ".join(docs) converts list into readable form
```
